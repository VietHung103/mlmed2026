```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv('mitbih_train.csv')
test_df = pd.read_csv('mitbih_test.csv')
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
X_train = train_df.iloc[:, :-1].values
y_train = train_df.iloc[:, -1].values.astype(int)
X_test = test_df.iloc[:, :-1].values
y_test = test_df.iloc[:, -1].values.astype(int)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
class_names = {0: 'Normal', 1: 'Supraventricular', 2: 'Ventricular', 3: 'Fusion', 4: 'Unknown'}
num_classes = len(class_names)
```

```python
num_classes
```

```python
unique, counts = np.unique(y_train, return_counts=True)
print("Training distribution:")
for cls, count in zip(unique, counts):
    print(f"  {class_names[cls]}: {count} ({count/len(y_train)*100:.1f}%)")
```

```python
plt.figure(figsize=(6, 4))
plt.bar([class_names[c] for c in unique], counts, color='green')
plt.title('Training Set Distribution')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("Training_Distribution.png")
plt.show()
```

```python
unique_test, counts_test = np.unique(y_test, return_counts=True)

plt.figure(figsize=(6, 4))
plt.bar([class_names[c] for c in unique_test], counts_test, color='purple')
plt.title('Test Set Distribution')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("Test_Distribution.png")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].bar([class_names[c] for c in unique], counts, color='green')
axes[0].set_title('Training Set Distribution')
axes[0].tick_params(axis='x', rotation=45)

unique_test, counts_test = np.unique(y_test, return_counts=True)
axes[1].bar([class_names[c] for c in unique_test], counts_test, color='purple')
axes[1].set_title('Test Set Distribution')
axes[1].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.savefig("Distribution.pdf")
plt.show()
```

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
log_reg = LogisticRegression(max_iter=1000)
svm = SVC(probability=True)
```

```python
ensemble_model = VotingClassifier(
    estimators=[
        ('lr', log_reg),
        ('svm', svm)
    ],
    voting='soft'  
)
```

```python
ensemble_model.fit(X_train, y_train)
```

```python
y_pred = ensemble_model.predict(X_test)
```

```python
print("Accuracy: ", accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5, weights='distance', metric='euclidean')
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
